```python
import pandas as pd

# 업로드된 파일 경로
file_path = '../data/jobs.csv'

# 파일 내용 확인
df = pd.read_csv(file_path)
df.head()
```

```python
# 근무내용에서 직무 내용 분리
df['업체명'] = df['근무내용'].str.extract(r'<(.+?)>')  # <> 안에 있는 업체명 추출
df['직무내용'] = df['근무내용'].str.split(r'\*직무내용:').str[1]  # '*직무내용:' 이후 텍스트 추출
df['직무내용'] = df['근무내용'].str.split(r'\*직무내용:').str[1]  # '*직무내용:' 이후 텍스트 추출

# 등록일에서 날짜 추출
df['등록일'] = df['등록일'].str.extract(r'등록일(\d{4}-\d{2}-\d{2})')

# 급여에서 금액 추출
df['급여'] = df['급여'].str.replace(r'급여', '').str.strip()

# 정리된 데이터프레임
df_cleaned = df[['구인상태', '업체명', '직종', '채용인원', '근무시간', '근무지역', '급여', '성별', '연령', '지원방법', '직무내용', '등록일']]
df_cleaned.head()
```

```python
# '직종', '근무지역' 등에서 특정 단어 제거
df_cleaned.loc[:, '직종'] = df_cleaned['직종'].str.replace('직종', '', regex=False).str.strip()
df_cleaned.loc[:, '근무지역'] = df_cleaned['근무지역'].str.replace('근무지역', '', regex=False).str.strip()
df_cleaned.loc[:, '구인상태'] = df_cleaned['구인상태'].str.replace('구인상태', '', regex=False).str.strip()
df_cleaned.loc[:, '지원방법'] = df_cleaned['지원방법'].str.replace('지원방법', '', regex=False).str.strip()
df_cleaned.loc[:, '연령'] = df_cleaned['연령'].str.replace('연령', '', regex=False).str.strip()
df_cleaned.loc[:, '성별'] = df_cleaned['성별'].str.replace('성별', '', regex=False).str.strip()

df_cleaned.head()
```

```python
output_file_path = '../data/cleaned_jobs.csv'
df_cleaned.to_csv(output_file_path, index=False)

output_file_path
```
